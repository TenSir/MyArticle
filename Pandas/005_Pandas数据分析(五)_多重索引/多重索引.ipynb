```python
import pandas as pd
import numpy as np

# 创建一个包含重复多级索引的DataFrame
data = np.random.randint(0, 100, size=(6, 2))
index_arrays = [
    ['a', 'a', 'b', 'b', 'c', 'c'],  # 第一级索引，存在重复
    [1, 2, 1, 2, 1, 2],               # 第二级索引，存在重复
    ['Name1', 'Name2', 'Name1', 'Name2', 'Name3', 'Name4']  # 第三级索引
]

multi_index = pd.MultiIndex.from_arrays(index_arrays, names=['Index1', 'Index2', 'Name'])
df = pd.DataFrame(data, columns=['Column1', 'Column2'])
df.index = multi_index

# 去除重复的多级索引
df_unique = df.drop_duplicates()

# 假设我们要选取第一行的数据，使用loc进行切片
# 注意：由于我们不确定去除重复后的索引顺序，我们可以使用索引值的标签进行选择
sliced_df = df_unique.loc[('a', 1), :]

# 打印结果查看
print(sliced_df)
```
